```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/QuantEcon/lecture-python-programming/master/source/_static/lecture_specific/pandas/data/test_pwt.csv')
type(df)
```

```python
df
```

```python
df = df[['country', 'POP', 'tcgdp']]
df
```

```python
df = df.set_index('country') #Removes the redundant numbers and indexes by country
df
```

```python
df.columns = 'Population', 'Total_GDP' #Renames the columns
df
```

```python
df['Population'] = df['Population'] * 1e3 #Converts to single units 1e3 essentially means 1 to the power of 3
df
```

```python
df['GDP_Per_Capita'] = df['Total_GDP'] * 1e6 / df['Population'] #1e6 because we're multiplying by a mill
df
```

```python
gdp_grph = df['GDP_Per_Capita'].plot(kind='bar')
gdp_grph.set_xlabel('country', fontsize=12)
gdp_grph.set_ylabel('GDP_Per_Capita', fontsize=12)
plt.show()
```

```python
df = df.sort_values(by='GDP_Per_Capita', ascending=False)
df
```

```python
gdp_grph = df['GDP_Per_Capita'].plot(kind='bar')
gdp_grph.set_xlabel('country', fontsize=12)
gdp_grph.set_ylabel('GDP_Per_Capita', fontsize=12)
plt.show()
```
